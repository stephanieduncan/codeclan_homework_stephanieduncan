```python
#obtaining the mean, maximum and minimum calories for drinks

max_drinks = drinks_content["Calories"].max()
max_drinks
```

```python
min_calories = drinks_content["Calories"].min()
min_calories
```

```python
mean_drinks = drinks_content["Calories"].mean()
mean_drinks
```

```python
#creating a data frame containing the drinks with the least and most calories 
drinks_content.loc[[drinks_content["Calories"].idxmin(), 
                    drinks_content["Calories"].idxmax()]].copy()

#Tazo Tea has zero calories

#Signature Espresso Drink, White Chocolate Mocha(without whipped cream) has the most calories
```

```python
#Creating a data frame consisting of select columns
drinks = drinks_content.loc[:, ["Beverage_category", "Beverage", "Beverage_prep", "Calories"]].copy()
drinks
```

```python
#checking for missing values in the new drinks data frame

drinks.isna().sum()

#there are no missing values
```

Creating a new variable named "calorie_diff" to store the difference between 135 calories (equivalent to 10 cubes of sugar) and the calories in each drink


```python
drinks.loc[:, "calorie_diff"] = 135 - drinks["Calories"]
drinks
```

Summaring the mean number of calroies in each beverage category

```python
mean_table = drinks.groupby("Beverage_category")["Calories"].mean().reset_index(name = "mean_calories")
mean_table
```

```python
#obtaining a view of the 3 categories of drinks with the most calories
mean_table = mean_table.sort_values("mean_calories", ascending = False)
mean_table.head(3)
```

The beverage categories with the highest mean number of calories for all drinks in that category are:

1. Smoothies
2. Frappuccino® Blended Coffee
3. Signature Espresso Drinks

```python
average_cals = drinks["Calories"].mean()
average_cals
```

```python
#The mean number of calories for all drinks is 194

#Showing which beverage_prep type contains more than the average calories of all drinks
drinks.loc[drinks.groupby("Beverage_prep")["Calories"].transform(lambda x: np.mean(x) > average_cals)]
```

```python
#Finding the best drink to consume if you are worried about consuming too many calories
drinks.loc[drinks["Beverage_category"] != "Tazo® Tea Drinks"].sort_values("Calories").head(1)
```

```python
#Filtering the dataset to include "Classic Espresson Drinks" only and daving as a new data frame name
espresso_drinks = drinks_content.loc[(drinks_content["Beverage_category"] == "Classic Espresso Drinks")].copy()
espresso_drinks
```

```python
#Grouping new data frame by beverage_prep then calculating the mean for each prep type
espresso_drinks.groupby("Beverage_prep").mean()
```

```python
#sorting by Calories in descending order
espresso_drinks.groupby("Beverage_prep").mean().sort_values("Calories", ascending = False)
```
